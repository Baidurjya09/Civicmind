## 🎮 GPU Configuration and Device Setup

This section will:
1. Check if CUDA/GPU is available on this runtime
2. Display GPU information (name, VRAM capacity) if available
3. Set the global device variable for training (cuda or cpu)
4. Configure mixed precision (fp16) training if GPU is available
5. Display warnings and recommendations if running on CPU

**Expected Runtime**: ~5 seconds

**Note**: GRPO training requires a GPU for reasonable performance (45 min on T4 GPU vs 10+ hours on CPU). Q-learning training works fine on CPU (10 seconds).

```python
import torch

print("\n" + "="*60)
print("🎮 GPU Configuration and Device Setup")
print("="*60 + "\n")

# Check CUDA availability
cuda_available = torch.cuda.is_available()

if cuda_available:
    # GPU is available
    device = torch.device("cuda")
    
    # Get GPU properties
    gpu_props = torch.cuda.get_device_properties(0)
    gpu_name = gpu_props.name
    gpu_vram_gb = gpu_props.total_memory / (1024**3)  # Convert bytes to GB
    
    print("✅ GPU DETECTED!\n")
    print(f"   GPU Name: {gpu_name}")
    print(f"   VRAM Capacity: {gpu_vram_gb:.2f} GB")
    print(f"   CUDA Version: {torch.version.cuda}")
    print(f"   Device: {device}\n")
    
    # Configure mixed precision (fp16) for faster training
    use_fp16 = True
    print("✅ Mixed Precision (fp16) Training: ENABLED")
    print("   This will speed up GRPO training and reduce memory usage.\n")
    
    # Display training time estimates
    print("📊 Estimated Training Times:")
    print(f"   - Q-Learning: ~10 seconds (CPU-based)")
    print(f"   - GRPO: ~45 minutes (GPU-accelerated)\n")
    
    print("🚀 You're all set for GPU-accelerated training!\n")
    
else:
    # No GPU available - CPU mode
    device = torch.device("cpu")
    use_fp16 = False
    
    print("⚠️  NO GPU DETECTED - Running in CPU Mode\n")
    print(f"   Device: {device}")
    print(f"   Mixed Precision: DISABLED (CPU only)\n")
    
    # Display warnings and recommendations
    print("⚠️  IMPORTANT WARNINGS:\n")
    print("   1. Q-Learning training will work fine on CPU (~10 seconds)")
    print("   2. GRPO training will be VERY SLOW on CPU (10+ hours vs 45 min on GPU)")
    print("   3. We recommend using 'Quick Q-Learning' mode for CPU-only environments\n")
    
    print("💡 To enable GPU in Google Colab:")
    print("   1. Go to Runtime > Change runtime type")
    print("   2. Select 'T4 GPU' or 'GPU' from Hardware accelerator dropdown")
    print("   3. Click Save and restart the runtime")
    print("   4. Re-run all cells from the beginning\n")
    
    print("📊 Estimated Training Times (CPU):")
    print(f"   - Q-Learning: ~10 seconds")
    print(f"   - GRPO: ~10+ hours (NOT RECOMMENDED)\n")

# Store configuration in global variables for use in training cells
DEVICE = device
USE_FP16 = use_fp16
HAS_GPU = cuda_available

print("="*60)
print("✅ GPU configuration complete!")
print(f"   Global device variable set: DEVICE = {DEVICE}")
print(f"   Mixed precision enabled: USE_FP16 = {USE_FP16}")
print(f"   GPU available: HAS_GPU = {HAS_GPU}")
print("="*60 + "\n")
```

## 📊 Dataset Generation

This section will:
1. Instantiate the DatasetGenerator with configurable parameters
2. Generate synthetic training samples (default: 500 samples, 70% good / 30% bad actions)

4. Save dataset to `training/civicmind_dataset.jsonl` in JSONL format

**Good Actions**: Welfare investment, trust building, crisis response

The generator will display a progress bar and update every 100 samples.

```python
import os
import errno
from pathlib import Path

print("\n" + "="*60)
print("📊 Dataset Generation")
print("="*60 + "\n")

try:
    # Import DatasetGenerator
    from training.data_generator import DatasetGenerator
    
    # Get dataset parameters from widget (or use defaults)
    n_samples = dataset_samples.value if 'dataset_samples' in dir() else 500
    good_ratio = 0.7  # 70% good actions, 30% bad actions
    
    print(f"📋 Dataset Configuration:")
    print(f"   Total Samples: {n_samples}")
    print(f"   Good Actions: {int(n_samples * good_ratio)} ({good_ratio:.0%})")
    print(f"   Bad Actions: {int(n_samples * (1 - good_ratio))} ({1-good_ratio:.0%})")
    print(f"   Agent Types: 6 (mayor, health_minister, finance_officer, police_chief, infrastructure_head, media_spokesperson)")
    print()
    
    # Check disk space before generation
    try:
        import shutil
        stat = shutil.disk_usage(".")
        free_gb = stat.free / (1024**3)
        
        if free_gb < 0.1:  # Less than 100MB free
            print("⚠️  WARNING: Low disk space detected!")
            print(f"   Free space: {free_gb:.2f} GB")
            print("   Dataset generation may fail if disk space is insufficient.")
            print()
    except Exception as e:
        # Disk space check failed, but continue anyway
        pass
    
    # Instantiate DatasetGenerator
    print("🚀 Initializing DatasetGenerator...\n")
    generator = DatasetGenerator(n_samples=n_samples, good_ratio=good_ratio)
    
    # Generate dataset with progress bar
    print("⏳ Generating training samples...")
    print("   This will take approximately 30 seconds for 500 samples.\n")
    
    samples = generator.generate_dataset()
    
    print("\n✅ Dataset generation complete!\n")
    
    # Save dataset to file
    output_path = "training/civicmind_dataset.jsonl"
    
    print(f"💾 Saving dataset to: {output_path}")
    
    try:
        # Create directory if it doesn't exist
        Path(output_path).parent.mkdir(parents=True, exist_ok=True)
        
        # Save dataset
        generator.save_dataset(output_path)
        
        # Verify file was created
        if os.path.exists(output_path):
            file_size_kb = os.path.getsize(output_path) / 1024
            print(f"   File size: {file_size_kb:.2f} KB")
            print()
        else:
            raise FileNotFoundError(f"Dataset file was not created: {output_path}")
            
    except PermissionError as e:
        print(f"\n❌ ERROR: Permission denied when writing to {output_path}")
        print(f"   Error: {e}")
        print(f"\n   Troubleshooting:")
        print(f"   1. Check that you have write permissions in the current directory")
        print(f"   2. If running in Colab, ensure the repository is in a writable location")
        print(f"   3. Try saving to a different directory (e.g., /tmp/)")
        raise
        
    except OSError as e:
        if e.errno == errno.ENOSPC:
            print(f"\n❌ ERROR: Insufficient disk space!")
            print(f"   Error: {e}")
            print(f"\n   Troubleshooting:")
            print(f"   1. Free up disk space by deleting temporary files")
            print(f"   2. Reduce the number of samples (use dataset_samples widget)")
            print(f"   3. If running in Colab, restart the runtime to clear temporary files")
            raise
        else:
            print(f"\n❌ ERROR: Failed to save dataset")
            print(f"   Error: {e}")
            raise
    
    # Display dataset statistics
    print("="*60)
    print("📈 Dataset Statistics")
    print("="*60 + "\n")
    
    stats = generator.get_statistics()
    
    print(f"Total Samples: {stats['total_samples']}")
    print(f"\nAction Distribution:")
    print(f"   Good Actions: {stats['good_samples']} ({stats['good_ratio']:.1%})")
    print(f"   Bad Actions:  {stats['bad_samples']} ({stats['bad_ratio']:.1%})")
    
    print(f"\nSamples per Agent:")
    for agent_id, count in stats['samples_per_agent'].items():
        percentage = (count / stats['total_samples']) * 100
        print(f"   {agent_id:25s}: {count:3d} samples ({percentage:.1f}%)")
    
    print("\n" + "="*60)
    print("✅ Dataset generation complete!")
    print(f"   Dataset saved to: {output_path}")
    print(f"   Total samples: {stats['total_samples']}")
    print(f"   Ready for training!")
    print("="*60 + "\n")
    
except ImportError as e:
    print(f"\n❌ ERROR: Failed to import DatasetGenerator")
    print(f"   Error: {e}")
    print(f"   Current directory: {os.getcwd()}")
    print(f"\n   Troubleshooting:")
    print(f"   1. Ensure you're in the CivicMind repository directory")
    print(f"   2. Check that training/data_generator.py exists")
    print(f"   3. Re-run the environment setup cells")
    raise
    
except Exception as e:
    print(f"\n❌ ERROR: Dataset generation failed")
    print(f"   Error type: {type(e).__name__}")
    print(f"   Error message: {e}")
    print(f"\n   Troubleshooting:")
    print(f"   1. Check the error message above for specific details")
    print(f"   2. Ensure all dependencies are installed (re-run setup cells)")
    print(f"   3. Try reducing the number of samples if memory is an issue")
    print(f"   4. Check that you have sufficient disk space")
    raise
```

## ⚡ Q-Learning Training

This section will:
1. Instantiate the QLearningTrainer with configurable parameters (default: 2000 episodes)
2. Train the tabular RL model using epsilon-greedy exploration with linear decay
3. Display training progress every 200 episodes (epsilon, states learned, average reward)
4. Save the trained Q-table to `training/checkpoints/rl_policy.pkl`
5. Display final statistics (total states learned, final epsilon, training time)
6. Generate training curve plot showing reward progression over episodes

**Expected Runtime**: ~10 seconds on CPU

**Training Details**:
- **Algorithm**: Q-learning with epsilon-greedy exploration
- **State Representation**: Discretized trust, GDP, survival, budget, crisis count, rebel status
- **Action Space**: 4-6 actions per agent (hold, invest_in_welfare, emergency_budget_release, etc.)
- **Epsilon Decay**: Linear from 1.0 (full exploration) to 0.1 (mostly exploitation)
- **Learning Rate**: 0.1 (alpha)
- **Discount Factor**: 0.95 (gamma)

The trainer will display progress updates every 200 episodes and save a checkpoint when training completes.

```python
import os
import pickle
from pathlib import Path

print("\n" + "="*70)
print("⚡ Q-Learning Training")
print("="*70 + "\n")

try:
    # Import QLearningTrainer
    from training.q_learning_trainer import QLearningTrainer
    
    # Get training parameters from widget (or use defaults)
    n_episodes = q_episodes.value if 'q_episodes' in dir() else 2000
    
    print(f"📋 Training Configuration:")
    print(f"   Episodes: {n_episodes}")
    print(f"   Epsilon: 1.0 → 0.1 (linear decay)")
    print(f"   Learning Rate: 0.1")
    print(f"   Discount Factor (gamma): 0.95")
    print(f"   Device: CPU (Q-learning is CPU-based)")
    print()
    
    # Instantiate QLearningTrainer with default parameters
    print("🚀 Initializing QLearningTrainer...\n")
    trainer = QLearningTrainer(
        episodes=n_episodes,
        epsilon_start=1.0,
        epsilon_end=0.1,
        learning_rate=0.1,
        gamma=0.95
    )
    
    print("⏳ Starting Q-learning training...")
    print("   This will take approximately 10 seconds for 2000 episodes.")
    print("   Progress updates will be displayed every 200 episodes.\n")
    
    # Train the model
    stats = trainer.train()
    
    print("\n✅ Training complete!\n")
    
    # Save checkpoint to training/checkpoints/rl_policy.pkl
    checkpoint_path = "training/checkpoints/rl_policy.pkl"
    
    print(f"💾 Saving checkpoint to: {checkpoint_path}")
    
    try:
        # Create checkpoint directory if it doesn't exist
        Path(checkpoint_path).parent.mkdir(parents=True, exist_ok=True)
        
        # Save checkpoint
        trainer.save_checkpoint(checkpoint_path)
        
        # Verify file was created
        if os.path.exists(checkpoint_path):
            file_size_kb = os.path.getsize(checkpoint_path) / 1024
            print(f"   File size: {file_size_kb:.2f} KB")
            print()
        else:
            raise FileNotFoundError(f"Checkpoint file was not created: {checkpoint_path}")
            
    except PermissionError as e:
        print(f"\n❌ ERROR: Permission denied when writing to {checkpoint_path}")
        print(f"   Error: {e}")
        print(f"\n   Troubleshooting:")
        print(f"   1. Check that you have write permissions in the current directory")
        print(f"   2. If running in Colab, ensure the repository is in a writable location")
        print(f"   3. Try saving to a different directory (e.g., /tmp/)")
        raise
        
    except OSError as e:
        if hasattr(e, 'errno') and e.errno == 28:  # ENOSPC
            print(f"\n❌ ERROR: Insufficient disk space!")
            print(f"   Error: {e}")
            print(f"\n   Troubleshooting:")
            print(f"   1. Free up disk space by deleting temporary files")
            print(f"   2. If running in Colab, restart the runtime to clear temporary files")
            raise
        else:
            print(f"\n❌ ERROR: Failed to save checkpoint")
            print(f"   Error: {e}")
            raise
    
    except pickle.PicklingError as e:
        print(f"\n❌ ERROR: Failed to serialize Q-table")
        print(f"   Error: {e}")
        print(f"\n   Troubleshooting:")
        print(f"   1. The Q-table may contain non-serializable objects")
        print(f"   2. Try restarting the training with a fresh trainer instance")
        raise
    
    # Display final statistics
    print("="*70)
    print("📊 Final Training Statistics")
    print("="*70 + "\n")
    
    print(f"Total Episodes: {stats['episodes']}")
    print(f"Total States Learned: {stats['states_learned']}")
    print(f"Final Epsilon: {stats['final_epsilon']:.3f}")
    print(f"Training Time: {stats['training_time']:.2f} seconds")
    
    # Calculate average reward for last 100 episodes
    if len(stats['episode_rewards']) >= 100:
        final_avg_reward = sum(stats['episode_rewards'][-100:]) / 100
        print(f"Final Avg Reward (last 100 episodes): {final_avg_reward:.4f}")
    
    print()
    
    # Generate training curve plot
    print("="*70)
    print("📈 Generating Training Curve Plot")
    print("="*70 + "\n")
    
    plot_path = "evidence/plots/q_learning_training_curve.png"
    
    try:
        # Create plots directory if it doesn't exist
        Path(plot_path).parent.mkdir(parents=True, exist_ok=True)
        
        # Generate and save plot (show=False for notebook compatibility)
        print(f"📊 Generating plot: {plot_path}")
        trainer.plot_training_curve(save_path=plot_path, show=False)
        
        # Display the plot in notebook
        try:
            from IPython.display import Image, display
            display(Image(filename=plot_path))
        except Exception as e:
            print(f"   ℹ️  Plot saved but could not be displayed: {e}")
            print(f"   You can view it at: {plot_path}")
        
        print()
        
    except ImportError as e:
        print(f"⚠️  WARNING: Could not generate plot (matplotlib not available)")
        print(f"   Error: {e}")
        print(f"   Training completed successfully, but plot generation was skipped.")
        print()
    
    except Exception as e:
        print(f"⚠️  WARNING: Plot generation failed")
        print(f"   Error: {e}")
        print(f"   Training completed successfully, but plot generation was skipped.")
        print()
    
    # Summary
    print("="*70)
    print("✅ Q-Learning Training Complete!")
    print("="*70)
    print(f"   Checkpoint saved to: {checkpoint_path}")
    print(f"   States learned: {stats['states_learned']}")
    print(f"   Training time: {stats['training_time']:.2f}s")
    if os.path.exists(plot_path):
        print(f"   Training curve saved to: {plot_path}")
    print(f"   Ready for evaluation!")
    print("="*70 + "\n")
    
    # Store trainer in global variable for later use
    Q_LEARNING_TRAINER = trainer
    Q_LEARNING_STATS = stats
    
except ImportError as e:
    print(f"\n❌ ERROR: Failed to import QLearningTrainer")
    print(f"   Error: {e}")
    print(f"   Current directory: {os.getcwd()}")
    print(f"\n   Troubleshooting:")
    print(f"   1. Ensure you're in the CivicMind repository directory")
    print(f"   2. Check that training/q_learning_trainer.py exists")
    print(f"   3. Re-run the environment setup cells")
    raise
    
except Exception as e:
    print(f"\n❌ ERROR: Q-learning training failed")
    print(f"   Error type: {type(e).__name__}")
    print(f"   Error message: {e}")
    print(f"\n   Troubleshooting:")
    print(f"   1. Check the error message above for specific details")
    print(f"   2. Ensure all dependencies are installed (re-run setup cells)")
    print(f"   3. Ensure the dataset generation cell completed successfully")
    print(f"   4. Check that you have sufficient disk space for checkpoints")
    print(f"   5. If the error persists, try reducing the number of episodes")
    raise
```

## 📊 Model Evaluation

This section will:
1. Instantiate the EvaluationEngine with configurable parameters (default: 5 episodes)
2. Evaluate the trained Q-learning policy against baseline policies
3. Compare random baseline, heuristic baseline, and trained policy
4. Measure metrics: mean reward, final reward, survival rate, trust score, rebel spawn rate
5. Calculate improvement percentages over baselines
6. Save evaluation results to `evaluation/artifacts/training_results.json`
7. Display comparison table showing all policies side-by-side

**Expected Runtime**: ~2 minutes for 5 episodes per policy

**Evaluation Details**:
- **Random Baseline**: Random action selection (no learning)
- **Heuristic Baseline**: Rule-based policy (hand-crafted rules)
- **Trained Policy**: Q-learning policy from previous training
- **Seeds**: Fixed seeds (42, 43, 44, 45, 46) for reproducibility

The evaluation will display progress for the first episode of each policy and aggregate statistics across all episodes.

```python
import json
from pathlib import Path

print("\n" + "="*70)
print("📊 Model Evaluation")
print("="*70 + "\n")

try:
    # Import EvaluationEngine and baseline policies
    from training.evaluation_engine import EvaluationEngine, random_policy, heuristic_policy
    
    # Get evaluation parameters from widget (or use defaults)
    n_eval_episodes = eval_episodes.value if 'eval_episodes' in dir() else 5
    
    print(f"📋 Evaluation Configuration:")
    print(f"   Episodes per policy: {n_eval_episodes}")
    print(f"   Max weeks per episode: 20")
    print(f"   Difficulty: 3")
    print(f"   Seeds: {[42 + i for i in range(n_eval_episodes)]}")
    print()
    
    # Instantiate EvaluationEngine
    print("🚀 Initializing EvaluationEngine...\n")
    engine = EvaluationEngine(
        n_episodes=n_eval_episodes,
        max_weeks=20,
        difficulty=3
    )
    
    # Create trained policy function from Q-learning trainer
    if 'Q_LEARNING_TRAINER' in dir():
        trained_policy_fn = Q_LEARNING_TRAINER.get_policy(epsilon=0.0)  # Greedy policy
        print("✅ Loaded trained Q-learning policy\n")
    else:
        print("⚠️  WARNING: Q-learning trainer not found!")
        print("   Attempting to load from checkpoint...\n")
        
        try:
            from training.q_learning_trainer import QLearningTrainer
            trainer = QLearningTrainer()
            trainer.load_checkpoint("training/checkpoints/rl_policy.pkl")
            trained_policy_fn = trainer.get_policy(epsilon=0.0)
            print("✅ Loaded trained policy from checkpoint\n")
        except Exception as e:
            print(f"❌ Failed to load checkpoint: {e}")
            print("   Evaluation will only compare baseline policies.\n")
            trained_policy_fn = None
    
    # Prepare policies for comparison
    policies = {
        "Random Baseline": random_policy,
        "Heuristic Baseline": heuristic_policy,
    }
    
    if trained_policy_fn is not None:
        policies["Trained Q-Learning"] = trained_policy_fn
    
    print("⏳ Starting evaluation...")
    print(f"   Evaluating {len(policies)} policies with {n_eval_episodes} episodes each.\n")
    
    # Compare all policies
    results = engine.compare_policies(policies)
    
    print("\n✅ Evaluation complete!\n")
    
    # Display comparison table
    print("="*70)
    print("📈 Policy Comparison")
    print("="*70 + "\n")
    
    print(f"{'Policy':<25} {'Mean Reward':>12} {'Final Reward':>13} {'Survival':>10} {'Rebel%':>8}")
    print("-" * 70)
    
    for policy_name, stats in results.items():
        print(
            f"{policy_name:<25} "
            f"{stats['mean_reward_avg']:>12.4f} "
            f"{stats['final_reward_avg']:>13.4f} "
            f"{stats['survival_avg']:>10.1%} "
            f"{stats['rebel_rate']:>8.1%}"
        )
    
    print()
    
    # Calculate and display improvements
    if "Random Baseline" in results and "Trained Q-Learning" in results:
        random_reward = results["Random Baseline"]["mean_reward_avg"]
        trained_reward = results["Trained Q-Learning"]["mean_reward_avg"]
        
        delta = trained_reward - random_reward
        pct = (delta / max(random_reward, 0.001)) * 100
        
        print("="*70)
        print("🎯 Improvement Analysis")
        print("="*70)
        print(f"\nTrained vs Random Baseline:")
        print(f"   Absolute improvement: {delta:+.4f}")
        print(f"   Relative improvement: {pct:+.1f}%")
        
        # Trust improvement
        random_trust = results["Random Baseline"]["trust_avg"]
        trained_trust = results["Trained Q-Learning"]["trust_avg"]
        trust_delta = trained_trust - random_trust
        trust_pct = (trust_delta / max(random_trust, 0.001)) * 100
        
        print(f"\nTrust Score Improvement:")
        print(f"   Absolute improvement: {trust_delta:+.4f}")
        print(f"   Relative improvement: {trust_pct:+.1f}%")
        print()
    
    if "Heuristic Baseline" in results and "Trained Q-Learning" in results:
        heuristic_reward = results["Heuristic Baseline"]["mean_reward_avg"]
        trained_reward = results["Trained Q-Learning"]["mean_reward_avg"]
        
        delta = trained_reward - heuristic_reward
        pct = (delta / max(heuristic_reward, 0.001)) * 100
        
        print(f"Trained vs Heuristic Baseline:")
        print(f"   Absolute improvement: {delta:+.4f}")
        print(f"   Relative improvement: {pct:+.1f}%")
        print()
    
    # Save results to JSON
    output_path = "evaluation/artifacts/training_results.json"
    
    print("="*70)
    print(f"💾 Saving evaluation results to: {output_path}")
    
    try:
        # Create directory if needed
        Path(output_path).parent.mkdir(parents=True, exist_ok=True)
        
        # Save results
        engine.save_results(results, output_path)
        
        # Verify file was created
        if os.path.exists(output_path):
            file_size_kb = os.path.getsize(output_path) / 1024
            print(f"   File size: {file_size_kb:.2f} KB")
        
    except Exception as e:
        print(f"⚠️  WARNING: Failed to save results: {e}")
        print("   Evaluation completed successfully, but results were not saved.")
    
    print("\n" + "="*70)
    print("✅ Model Evaluation Complete!")
    print("="*70)
    print(f"   Policies evaluated: {len(results)}")
    print(f"   Episodes per policy: {n_eval_episodes}")
    if os.path.exists(output_path):
        print(f"   Results saved to: {output_path}")
    print(f"   Ready for artifact export!")
    print("="*70 + "\n")
    
    # Store results in global variable
    EVALUATION_RESULTS = results
    
except ImportError as e:
    print(f"\n❌ ERROR: Failed to import EvaluationEngine")
    print(f"   Error: {e}")
    print(f"   Current directory: {os.getcwd()}")
    print(f"\n   Troubleshooting:")
    print(f"   1. Ensure you're in the CivicMind repository directory")
    print(f"   2. Check that training/evaluation_engine.py exists")
    print(f"   3. Re-run the environment setup cells")
    raise
    
except Exception as e:
    print(f"\n❌ ERROR: Model evaluation failed")
    print(f"   Error type: {type(e).__name__}")
    print(f"   Error message: {e}")
    print(f"\n   Troubleshooting:")
    print(f"   1. Check the error message above for specific details")
    print(f"   2. Ensure Q-learning training completed successfully")
    print(f"   3. Check that the checkpoint file exists")
    print(f"   4. Ensure all dependencies are installed")
    raise
```

3. Create a timestamped zip archive (e.g., `civicmind_results_20260425_143022.zip`)

```python
from pathlib import Path

print("\n" + "="*70)
print("📦 Artifact Export and Download")
print("="*70 + "\n")

try:
    # Import ArtifactExporter
    from training.artifact_exporter import ArtifactExporter
    
    print("🚀 Initializing ArtifactExporter...\n")
    exporter = ArtifactExporter()
    
    # Collect all artifacts
    print("📂 Collecting artifacts from:")
    print("   - training/checkpoints/")
    print("   - evaluation/artifacts/")
    print("   - evidence/eval/")
    print("   - evidence/plots/")
    print()
    
    artifacts = exporter.collect_artifacts()
    
    if not artifacts:
        print("\n⚠️  WARNING: No artifacts found to package!")
        print("   This may indicate that training or evaluation did not complete successfully.")
        print("   Please ensure all previous cells have been executed.")
    else:
        print(f"\n✅ Collected {len(artifacts)} files\n")
        
        # Create archive
        print("📦 Creating zip archive...\n")
        archive_path = exporter.create_archive()
        
        # Get archive information
        print("\n" + "="*70)
        print("📊 Archive Information")
        print("="*70 + "\n")
        
        info = exporter.get_archive_info()
        
        print(f"Archive Path: {info['path']}")
        print(f"Archive Size: {info['size_mb']} MB")
        print(f"Total Files: {info['file_count']}")
        print()
        
        # Display contents summary
        print("Contents Summary:")
        
        # Group by directory
        by_dir = {}
        for item in info['contents']:
            dir_name = str(Path(item['filename']).parent)
            if dir_name not in by_dir:
                by_dir[dir_name] = []
            by_dir[dir_name].append(item)
        
        for dir_name, files in sorted(by_dir.items()):
            total_size = sum(f['size_kb'] for f in files)
            print(f"   {dir_name}/ ({len(files)} files, {total_size:.1f} KB)")
        
        print()
        
        # Trigger download
        print("="*70)
        print("📥 Triggering Download")
        print("="*70 + "\n")
        
        try:
            exporter.trigger_download()
        except Exception as e:
            print(f"⚠️  WARNING: Download trigger failed: {e}")
            print(f"   You can manually download the file from: {archive_path}")
        
        print("\n" + "="*70)
        print("✅ Artifact Export Complete!")
        print("="*70)
        print(f"   Archive: {archive_path}")
        print(f"   Size: {info['size_mb']} MB")
        print(f"   Files: {info['file_count']}")
        print(f"   Ready for submission!")
        print("="*70 + "\n")
        
        # Store archive path in global variable
        ARCHIVE_PATH = archive_path

except ImportError as e:
    print(f"\n❌ ERROR: Failed to import ArtifactExporter")
    print(f"   Error: {e}")
    print(f"   Current directory: {os.getcwd()}")
    print(f"\n   Troubleshooting:")
    print(f"   1. Ensure you're in the CivicMind repository directory")
    print(f"   2. Check that training/artifact_exporter.py exists")
    print(f"   3. Re-run the environment setup cells")
    raise
    
except Exception as e:
    print(f"\n❌ ERROR: Artifact export failed")
    print(f"   Error type: {type(e).__name__}")
    print(f"   Error message: {e}")
    print(f"\n   Troubleshooting:")
    print(f"   1. Check the error message above for specific details")
    print(f"   2. Ensure previous cells completed successfully")
    print(f"   3. Check that you have sufficient disk space")
    print(f"   4. Verify write permissions in the current directory")
    raise
```
